```python
# season wise winner
season_match=matches.drop_duplicates(subset='season',keep='last')[['season','winner']].reset_index(drop=True)
season_match.sort_values('season')
```

```python
#toss winner is also a match winner
matches['toss_winner_is_winner'] = 'Lose'
matches['toss_winner_is_winner'].loc[matches.toss_winner == matches.winner] = 'Win'
temp=matches.toss_winner_is_winner.value_counts()
plt.pie(np.array(temp/temp.sum()*100),labels=temp.index,colors=['green','red'],shadow=True,startangle=90,autopct='%1.1f%%')
plt.title('percentage of toss winner to be a winner')
plt.show()
```

```python
matches.loc[matches.toss_winner == matches.winner]
```

```python
# win percentage who opted for batting first
no_of_wins=(matches.win_by_runs>0).sum()
no_of_loss=(matches.win_by_runs==0).sum()
total=float(no_of_wins+no_of_loss)
sizes=[(no_of_wins/total)*100,(no_of_loss/total)*100]
plt.pie(sizes,labels=['WIN','Loss'],colors=['green','red'],shadow=True,startangle=90,autopct='%1.1f%%')
plt.title('Chances of wining if opting batting first')
plt.show()
```

```python
#Top players
player_of_the_match=matches.player_of_match.value_counts()[:18]
```

```python
#let's see top players
temp_series = matches.player_of_match.value_counts()[:18]
labels = np.array(temp_series.index)
ind = np.arange(len(labels))
width = 0.8
fig, ax = plt.subplots()
rects = ax.bar(ind, np.array(temp_series), width=width)
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Player of the match awardees")

# create a function for labeling #
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1*height,
                '%d' % int(height),
                ha='center', va='bottom')
autolabel(rects)        

plt.show()
```
